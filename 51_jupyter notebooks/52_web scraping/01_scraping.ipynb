<前提知識>

・「Requests」はHTTPライブラリでWebページを取得するためのもの

・「Selenium」はブラウザ操作をするためのもの

・「Beautiful Soup」は取得したHTMLから情報を抽出するためのもの

------------[Reqests] の基本的な操作------------

```python
import requests

r = requests.get('https://news.yahoo.co.jp')  #get関数を用いてURLを指定→情報を格納する
```

```python
print(r.headers)  #headers = webサイトの「header」情報を取得
print("--------")
print(r.encoding) #encoding = webサイトの「文字コード」情報を取得
#print(r.content)   #content = webサイトの「bode」以下の情報を取得
```

------------[Beautiful Soup] の活用------------

```python
from bs4 import BeautifulSoup
 
html = "<h1>sayhello</h1>,<h1>saysay</h1>,<h2>say</h2>"
 
soup = BeautifulSoup(html, "html.parser")  #第一引数：解析したいもの、第二引数：パーサ
 
print(soup.select("h1"))
```

------------[Beautiful Soup / Reqests] の活用------------

```python
import requests
from bs4 import BeautifulSoup
 
r = requests.get("https://news.yahoo.co.jp/")
 
soup = BeautifulSoup(r.content, "html.parser")
 
# ニュース一覧を抽出
print(soup.find("ul", "newsFeed_list"))
```

```python
import requests
from bs4 import BeautifulSoup
	
r = requests.get("https://news.yahoo.co.jp/")
	
soup = BeautifulSoup(r.content, "html.parser")
	
#ニュース一覧のテキストのみ抽出
print(soup.find("ul", "newsFeed_list").text)
```
